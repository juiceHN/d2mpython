```python
#Rename a column
fhand.rename(columns={'Sub-Category':'SubCategory'},inplace=True)
fhand.head()
```

```python
#Average quantities sold per subcategory
fhand.SubCategory.unique()
qty_mean_subcat = fhand.groupby('SubCategory').Quantity.mean()
print(type(qty_mean_subcat))
qty_mean_subcat.sort_values(ascending=False)
```

```python
#Average sales per subcategory
fhand.SubCategory.unique()
sales_mean_subcat = fhand.groupby('SubCategory').Sales.mean()
print(type(sales_mean_subcat))
sales_mean_subcat.sort_values(ascending=False)

#In terms of quantity sold, Fasteners came first 
#but are the last in terms of sales (=/= Table for ex)
#Their price must be really low = could be improved ==> Increase its price
#Same for Labels?

#Different results compared with Tableau.......???!!!!
```

```python
#Average discount per subcategory
fhand.SubCategory.unique()
discount_mean_subcat = fhand.groupby('SubCategory').Discount.mean()
print(type(discount_mean_subcat))
pourcentage_discount_mean_subcat = discount_mean_subcat*100
pourcentage_discount_mean_subcat.sort_values(ascending=False)

#Thanks to this rank we now undertand why many Binders are sold (qty)
#While its sales aren't that high. Big discounts!
```

```python
#Visualize the repartition of sales per Subcategory
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
plt.figure(figsize=(16,7))
sb.barplot(data=fhand, x='SubCategory', y='Sales',ci=None)
#None removes error bars

#Why not same results as on Tableau?? Here best Sales are Copiers and Machines vs Tableau : Phones and Chairs!!!
```

```python
#Adding the figures on the plot
plt.figure(figsize=(16,7))
chart=sb.barplot(data=fhand,x='SubCategory',y='Sales',ci=None)
def values_on_bars(chart):
    for p in chart.patches:
        chart.annonate(format(p.get_height(),'.2f'),
                      (p.get_x()+p.get_width()/2.,
                      p.get_height()),
                      ha='center',
                      va='center',
                      xytext=(0, 10),
                      textcoords='offset points')
values_on_bars(chart)
#It does not work????
```

```python
#Sales evolution per year
plt.figure(figsize=(16,7))
chart=sb.lineplot(data=fhand, x='Order Year', y='Sales')
chart.set_title('Total sales evolution per year')
chart.set_xlabel('Year')
chart.set_ylabel('Total Sales')
```

```python
#Profit evolution per year per Category
plt.figure(figsize=(16,7))
chart=sb.lineplot(data=fhand, x='Order Year', y='Profit', hue='Category')
chart.set_title('Total profit evolution per year per Category')
chart.set_xlabel('Year')
chart.set_ylabel('Total Profit')
```

```python
#Yearly sales evolution per Category
by_year_cat_sales = fhand.groupby(['Order Year','Category']).Sales.sum().reset_index()

plt.figure(figsize=(10,7))
chart=sb.lineplot(data=by_year_cat_sales, x='Order Year',hue= 'Category',y='Sales',ci=None)
chart.set_title('Total Sales evolution per year per Category')
chart.set_xlabel('Year')
chart.set_ylabel('Total Sales')
```

```python
#Yearly sales evolution per Segment
by_year_segment_sales = fhand.groupby(['Order Year','Segment']).Sales.sum().reset_index()

plt.figure(figsize=(10,7))
chart=sb.lineplot(data=by_year_segment_sales, x='Order Year',hue= 'Segment',y='Sales',ci=None)
chart.set_title('Total Sales evolution per year per Segment')
chart.set_xlabel('Year')
chart.set_ylabel('Total Sales')
```

```python
#Yearly sales evolution per Region
by_year_region_sales = fhand.groupby(['Order Year','Region']).Sales.sum().reset_index()

plt.figure(figsize=(10,7))
chart=sb.lineplot(data=by_year_region_sales, x='Order Year',hue= 'Region',y='Sales',ci=None)
chart.set_title('Total Sales evolution per year per Region')
chart.set_xlabel('Year')
chart.set_ylabel('Total Sales')
```

```python
#Boxplot Discount per Subcategory
plt.figure(figsize=(16,6))
chart=sb.boxplot(data=fhand,x='SubCategory',y='Discount')
chart.set_title('Boxplot of Discount per Subcategory')
```

```python
#How many days between Order Date and Ship Date
from datetime import datetime

fhand['Order Date'] = pd.to_datetime(fhand['Order Date'])
fhand['Ship Date'] = pd.to_datetime(fhand['Ship Date'])

fhand['Order Date'] = datetime.strptime(fhand['Order Date'], "%Y-%m-%d")
fhand['Ship Date'] = datetime.strptime(fhand['Ship Date'], "%Y-%m-%d")

d1=fhand['Order Date']
d2=fhand['Ship Date']

return abs((d1-d2).days)
```
