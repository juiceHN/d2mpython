```python
def values_on_bars(chart):
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.2f'), 
                       (p.get_x() + p.get_width() / 2., 
                       p.get_height()), 
                       ha = 'center', 
                       va = 'center',
                       xytext = (0, 10), 
                       textcoords = 'offset points')
```

```python
plt.figure( figsize = (10, 7) )
chart=sb.barplot(data=total_profit_region, x='Region',y='profit_margin', ci=None)
values_on_bars(chart)
plt.show()
```

```python
plt.figure( figsize = (10, 7) )
chart=sb.barplot(data=total_profit_region, x='Region',y='profit_margin', ci=None)
chart.set_title('Profit Margin per Region')
chart.set_xlabel('Region')
chart.set_ylabel('Profit/Sale (%)')
values_on_bars(chart)
plt.show()
```
